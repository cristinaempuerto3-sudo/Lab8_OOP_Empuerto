```python
#Step 1: Define a base class
class Shape:
    def area(self):
        return "Calculating area..."
    
#Step 2: Define subclasses with their implementations
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2
    
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

#Step 3: Create objects and call methods
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Circle area: {circle.area()}")
print(f"Rectangle area: {rectangle.area()}")
```

```python
# Step 1: Define Vector class
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Addition
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    # Subtraction
    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

# Step 2: Create objects
v1 = Vector(2, 3)
v2 = Vector(1, 4)

# Step 3: Perform operations
add_result = v1 + v2
sub_result = v1 - v2

# Step 4: Display results
print("Addition:", add_result.x, add_result.y)
print("Subtraction:", sub_result.x, sub_result.y)
```

```python
# Step 1: Define an abstract class
from abc import ABC, abstractmethod

class Appliance(ABC):
    @abstractmethod
    def turn_on(self):
        pass

# Step 2: Define a subclass implementing the abstract method
class WashingMachine(Appliance):
    def turn_on(self):
        return "Washing machine is now running."

# Step 3: Create an object of the subclass and call the method
wm = WashingMachine()
print(wm.turn_on())
```

```python
# Step 1: Define multiple classes
class A:
    def show(self):
        return "Class A"
class B(A):
    def show(self):
        return "Class B"
class C(A):
    def show(self):
        return "Class C"
class D(B, C):
    pass # Inherits show() based on MRO
# Step 2: Create an object of class D and call the method
obj = D()
print(obj.show()) # Output depends on MRO
```

```python
# Step 1: Define classes
class A:
    def show(self):
        return "This is Class A"

class B(A):
    def show(self):
        return "This is Class B"

class C(A):
    def show(self):
        return "This is Class C"

class D(B, C):
    pass

# Step 2: Create object of D
obj = D()

# Step 3: Call method (uses MRO)
print(obj.show())

# Step 4: Show Method Resolution Order
print(D.mro())
```

```python
#14
class ATM: 
    def __init__(self, balance): 
        self.balance = balance 

    def withdraw(self, amount): 
        if amount > self.balance: 
            return "Insufficient balance" 
        self.balance -= amount 
        return f"Withdrawn {amount}, Remaining balance: {self.balance}" 

atm = ATM(1000) 
print(atm.withdraw(200)) 
```
