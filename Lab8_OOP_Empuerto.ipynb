```python
#Step 1: Define a class named Car
class Car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

    def display_info(self):
        return f"{self.year} {self.brand} {self.model}"

#Step 2: Create an object of the Car class
my_car = Car("Toyota", "Corolla", 2022)

#Step 3: Print the car details using the method
print(my_car.display_info())
```

```python
#Try it
class Laptop:
    def __init__(self, brand, processor, ram):
        self.brand = brand
        self.processor = processor
        self.ram = ram

    def display_info(self):
        return f"{self.brand} {self.processor} {self.ram}GB"

#Step 2: Create an object of the Laptop class
my_laptop = Laptop("Lenovo", "Intel Core i3", 8)

#Step 3: Print the laptop details using the method
print(my_laptop.display_info())
```
