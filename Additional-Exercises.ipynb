```python
#Additional Exercise 1
class Book:
    def __init__(self, title, author, available):
        self.title = title.strip()
        self.author = author.strip()
        self.available = available


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author, available):
        self.books.append(Book(title, author, available))

    def find_book(self, title):
        title = title.strip().lower()
        for book in self.books:
            if book.title.lower() == title:
                return book
        return None

    def borrow_book(self, title):
        book = self.find_book(title)

        if not book:
            return "Book does not exist."

        if not book.available:
            return "Book is already borrowed."

        book.available = False
        return f"You successfully borrowed '{book.title}'."

    def return_book(self, title):
        book = self.find_book(title)

        if not book:
            return "Book does not exist."

        if book.available:
            return "Book was not borrowed."

        book.available = True
        return f"You successfully returned '{book.title}'."

    def display_books(self):
        print("\nLibrary Books:")
        for book in self.books:
            status = "Available" if book.available else "Borrowed"
            print(f"{book.title} by {book.author} - {status}")


# main program
def main():
    library = Library()

    # preloaded books
    library.add_book("The Trial", "Franz Kafka", True)
    library.add_book("Still Me", "Jojo Moyes", False)
    library.add_book("Goldilocks", "Laura Lam", True)
    library.add_book("The Hobbit", "J.R.R. Tolkien", True)

    # menu
    while True:
        action = input(
            "\nChoose (add / borrow / return / display / exit): "
        ).strip().lower()

        # add books
        if action == "add":
            try:
                n = int(input("How many books do you want to add? "))
                if n <= 0:
                    print("Must add at least 1 book.")
                    continue
            except:
                print("Invalid number.")
                continue

            added_books = []

            for _ in range(n):  
                title = input("Enter title: ")
                author = input("Enter author: ")

                status = input("Is it available? (yes/no): ").strip().lower()
                available = True if status == "yes" else False

                library.add_book(title, author, available)

                added_books.append(f"'{title}' by {author}")

            print("\nBooks successfully added:")
            for b in added_books:
                print("-", b)

        # borrow book
        elif action == "borrow":
            title = input("Enter book title to borrow: ")
            print(library.borrow_book(title))

        # return book
        elif action == "return":
            title = input("Enter book title to return: ")
            print(library.return_book(title))

        # display books
        elif action == "display":
            library.display_books()

        # exit program
        elif action == "exit":
            break

        else:
            print("Invalid action.")

    print("\nFinal library status:")
    library.display_books()


if __name__ == "__main__":
    main()
```

```python
#2
class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name.strip()
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def compute_pay(self):
        overtime_hours = 0
        base_hours = self.hours_worked

        if self.hours_worked > 160:
            overtime_hours = self.hours_worked - 160
            base_hours = 160

        base_salary = base_hours * self.hourly_rate
        overtime_pay = overtime_hours * self.hourly_rate * 1.5
        total_salary = base_salary + overtime_pay

        return base_salary, overtime_pay, total_salary


class PayrollSystem:
    def __init__(self):
        self.employees = []

    def add_employee(self, name, rate, hours):
        if rate < 0 or hours < 0:
            print("Invalid input: wage and hours must not be negative.")
            return

        self.employees.append(Employee(name, rate, hours))

    def display_payroll(self):
        if len(self.employees) == 0:
            print("No employees found.")
            return

        print("\n--- Payroll Summary ---")

        for emp in self.employees:
            base, overtime, total = emp.compute_pay()

            print(f"\nName: {emp.name}")
            print(f"Hours Worked: {emp.hours_worked:,.2f}")
            print(f"Base Salary: ₱{base:,.2f}")
            print(f"Overtime Pay: ₱{overtime:,.2f}")
            print(f"Total Salary: ₱{total:,.2f}")


# main program
payroll = PayrollSystem()

try:
    n = int(input("Enter number of employees: "))
    if n <= 0:
        print("You must enter at least one employee.")
        exit()
except:
    print("Invalid input. Please enter a number.")
    exit()

for i in range(n):
    print(f"\nEmployee {i+1}")

    name = input("Enter name: ")

    try:
        rate = float(input("Enter hourly wage: "))
        hours = float(input("Enter hours worked: "))
    except:
        print("Invalid input. Skipping employee.")
        continue

    payroll.add_employee(name, rate, hours)

payroll.display_payroll()
```

```python
#3
class Product:
    def __init__(self, name, price, stock):
        self.name = name.strip()
        self.price = price
        self.stock = stock


class CartItem:
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity


class ShoppingCart:
    def __init__(self):
        self.products = []
        self.cart = []

    def add_product(self, name, price, stock):
        self.products.append(Product(name, price, stock))

    def find_product(self, name):
        name = name.strip().lower()
        for p in self.products:
            if p.name.lower() == name:
                return p
        return None

    def add_to_cart(self, name, quantity):
        product = self.find_product(name)

        if not product:
            return "Invalid product selection."

        if quantity <= 0:
            return "Quantity must be greater than 0."

        if quantity > product.stock:
            return f"Not enough stock for {product.name}."

        # reduce stock
        product.stock -= quantity

        for item in self.cart:
            if item.product.name.lower() == product.name.lower():
                item.quantity += quantity
                return f"Added {quantity} more {product.name}(s) to cart."

        self.cart.append(CartItem(product, quantity))
        return f"{product.name} added to cart."

    def remove_from_cart(self, name, quantity):
        name = name.strip().lower()

        for item in self.cart:
            if item.product.name.lower() == name:
                if quantity <= 0:
                    return "Invalid quantity."

                if quantity > item.quantity:
                    return "You are removing more than what is in cart."

                item.quantity -= quantity
                item.product.stock += quantity

                if item.quantity == 0:
                    self.cart.remove(item)

                return f"Removed {quantity} {item.product.name}(s) from cart."

        return "Item not found in cart."

    # calculate total
    def get_total(self):
        total = 0
        for item in self.cart:
            total += item.product.price * item.quantity
        return total

    #display cart
    def display_cart(self):
        if not self.cart:
            print("\nCart is empty.")
            return

        print("\n--- CART CONTENTS ---")
        for item in self.cart:
            subtotal = item.product.price * item.quantity
            print(
                f"{item.product.name} x{item.quantity} "
                f"- ₱{subtotal:,.2f}"
            )

        print(f"\nTOTAL: ₱{self.get_total():,.2f}")


#main program
def main():
    shop = ShoppingCart()

    # predefined products
    shop.add_product("Kindle Paperwhite", 7500, 100)
    shop.add_product("Kobo Clara BW", 9500, 121)
    shop.add_product("Kobo Clara Color", 12000, 50)
    shop.add_product("Tumbler", 350, 50)
    shop.add_product("Pomodoro Timer", 1250, 3)
    shop.add_product("xteink x3", 3400, 10)
    shop.add_product("Vintage Pocket Watch", 5000, 5)
    shop.add_product("The Little Prince Paperback", 200, 100)
    shop.add_product("Uni Kurutoga KS", 270, 57)
    shop.add_product("Cacaolat 1L", 275, 176)

    while True:
        action = input(
            "\nChoose (add / remove / view / checkout / exit): "
        ).strip().lower()

        # add to cart
        if action == "add":
            name = input("Enter product name: ")
            try:
                qty = int(input("Enter quantity: "))
            except:
                print("Invalid quantity.")
                continue

            print(shop.add_to_cart(name, qty))

        # remove from cart
        elif action == "remove":
            name = input("Enter product name: ")
            try:
                qty = int(input("Enter quantity to remove: "))
            except:
                print("Invalid quantity.")
                continue

            print(shop.remove_from_cart(name, qty))

        # view cart
        elif action == "view":
            shop.display_cart()

        # checkout
        elif action == "checkout":
            print("\n--- FINAL RECEIPT ---")
            shop.display_cart()
            print("\nThank you for shopping!")
            break

        # exit
        elif action == "exit":
            break

        else:
            print("Invalid action.")


if __name__ == "__main__":
    main()
```
